## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create Bins
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize data Using pd.cut 
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins, labels=group_names)
purchase_data

# Use .groupby On Age Group 
age_grouped = purchase_data.groupby("Age Group")

#Calculate Total and Percentage
pur_count = age_grouped["SN"].nunique()
player_percent = (pur_count/tot_player)*100

# Create a Data Frame
age_demographics = pd.DataFrame({"Percentage of players":player_percent,"Total count":pur_count})

# Change Index to None And Display Data Frame
age_demographics.index.name = None
age_demographics
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Use .loc Function to Retrieve the Item ID, Item Name, and Item Price columns 
popular_data = purchase_data.loc[:,["Item ID","Item Name","Price"]]

#Group by Item ID and Item Name. 
popular_group = purchase_data.groupby(["Item ID","Item Name"])

#Calculate Purchase Count, Average Purchase Price,& Total Purchase Value
pur_count = popular_group["Age"].count()
avg_purchase = popular_group["Price"].mean()
total_purchase_value = popular_group["Price"].sum()

#Create Data Frame 
most_popular = pd.DataFrame({"Purchase count":pur_count,"Average Purchase Price":avg_purchase,
                                   "Total Purchase value":total_purchase_value})

#Sort The Purchase Count Column In Descending Order
most_popular = most_popular.sort_values("Purchase count", ascending = False)

# Display Data
most_popular.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Group by Item ID and Item Name. 
popular_group = purchase_data.groupby(["Item ID","Item Name"])

#Calculate Purchase Count, Average Purchase Price,& Total Purchase Value
pur_count = popular_group["Age"].count()
avg_purchase = popular_group["Price"].mean()
total_purchase_value = popular_group["Price"].sum()

#Create Data Frame
most_popular = pd.DataFrame({"Purchase count":pur_count,"Average Purchase Price":avg_purchase,
                                   "Total Purchase value":total_purchase_value})


#Sort The Total Purchase Value Column In Descending Order
most_popular = most_popular.sort_values("Total Purchase value", ascending = False)
most_popular["Total Purchase value"]  = most_popular["Total Purchase value"].map("${:,.2f}".format)
most_popular["Average Purchase Price"] = most_popular["Average Purchase Price"].map("${:.2f}".format)

# display the Most Profitable Items summary table 
most_popular.head()
```
